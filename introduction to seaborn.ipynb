# Seaborn : A Good Visualization Package in Python

Seaborn is a good visualization tool.

We could use it to explore the datasets or make some parameters chosen directly by clients and let clients to explore data by themselves.

More detail, please look at [seaborn tutorial](http://web.stanford.edu/~mwaskom/software/seaborn/index.html).

```python
###  import some packages

import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.get_dataset_names()
```

### Section 1: Visualizing the Distribution of a Dataset

Here, our aim is to explore the continuous variables. 

When we have one, two or more continous variables, how could we do in seaborn?

First, let us load a dataset - iris, for an example, and simply explore this dataset.

```python
iris = sns.load_dataset("iris")
```

```python
type(iris)
```

```python
iris.head()
```

```python
iris.describe()
```

```python
iris.columns
```

#### Section 1.1: look at a univariate distribution 

Here, we have a variable which takes continous values, we could look at its distribution.

Seaborn will draw a histogram and fit a kernel density estimate by default.

```python
sns.distplot(iris.sepal_length, rug = True, fit = stats.gausshyper);
```

#### Section 1.2: visualize a bivariate distribution of two variables

When there are two variables take continuous values, we could see their relationships by scatter plot.

```python
sns.jointplot(x = "sepal_length", y = "petal_length", data = iris);
```

```python
#  this plot works best with relatively large datasets
with sns.axes_style("white"):
    sns.jointplot(x = "sepal_length", y = "petal_length", data = iris, kind = "hex", color  ="k");
```

```python
sns.jointplot(x = "sepal_length", y = "petal_length", data = iris, kind = "kde");
```

```python
g = sns.jointplot(x = "sepal_length", y = "petal_length", data = iris, kind = "kde", color = "m")
g.plot_joint(plt.scatter, c = "w", s = 30, linewidth = 1, marker = "+")
g.ax_joint.collections[0].set_alpha(0);
```

#### Section 1.3: plot multiple pairwise bivariate distributions in a dataset

If we have more than two variables takes continuous values, we could see their relationships all at once.

```python
sns.pairplot(iris);
```

### Section 2: Visualizing Linear Relationships

Here, we want to visualization some relationships such as linear or curves.

So, there are two or more variables will be refered.

The response variable or object variable could takes continuous values or discrete values.

```python
tips = sns.load_dataset("tips")
```

```python
type(tips)
```

```python
tips.head()
```

```python
tips.describe()
```

#### Section 2.1 some simple relationships plots

Here, the response variable is "tip", predictor variable is "total_bill".

These two variables are all take continuous values.

This plot simply fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips);
```

```python
# predictor variable is size which takes discrete values
sns.lmplot(x = "size", y = "tip", data = tips, x_jitter = .05);
```

```python
# this method is good,it demonstrates the effect of different sizes to tips.
sns.lmplot(x = "size", y = "tip", data = tips, x_estimator = np.mean);
```

#### Section 2.2: fit many models directly and display the effects by seaborn

```python
anscombe = sns.load_dataset("anscombe")
```

```python
## Q1: the relationship is a curve, not a linear.
sns.lmplot(x = "x", y = "y", data = anscombe.query("dataset == 'II'"),
           ci = None, scatter_kws = {"s": 80});
```

```python
## Amazing! we could tune the parameter to fit the curve.
sns.lmplot(x = "x", y = "y", data = anscombe.query("dataset == 'II'"),
           order = 2, ci = None, scatter_kws = {"s": 80});
```

```python
## Q2: there is a outlier point, which has a bad effect to the fitting line.
sns.lmplot(x = "x", y = "y", data = anscombe.query("dataset == 'III'"),
           ci = None, scatter_kws = {"s": 80});
```

```python
## Good. we could fit a robust regression, which uses a different loss function to downweight relatively large residuals
sns.lmplot(x = "x", y = "y", data = anscombe.query("dataset == 'III'"),
           robust = True, ci = None, scatter_kws = {"s": 80});
```

```python
## Q3: there is a discrete response variable, what can we do?
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x = "total_bill", y = "big_tip", data = tips,
           y_jitter = .03);
```

```python
## We could use logistic regression !
sns.lmplot(x = "total_bill", y = "big_tip", data = tips,
           logistic = True, y_jitter = .03);
```

```python
## A greate method to nonlinear relationship between variables is lowess smoother.
sns.lmplot(x="total_bill", y="tip", data=tips,
           lowess=True);
```

#### Section 2.3: conditioning on other variables 

Here is a interesting part: explore the relationships between vaiables in deeper width.

If there is a discrete variable, when this variable chose different values, what will happen to the relationships between two other variables ?

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", data = tips);
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", data = tips, 
           markers = ["o", "x"], palette = "Set1");
```

```python
## add a discrete vaiable again
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "time", data = tips);
```

```python
## add two discrete vaiables again
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker",
           col = "time", row = "sex", data = tips);
```

```python
## control the size and shape of the plot
sns.lmplot(x = "total_bill", y = "tip", col = "day", data = tips,
           col_wrap = 2, size = 3);
```

```python
sns.lmplot(x = "total_bill", y = "tip", col = "day", data = tips,
           aspect = .5);
```

```python
## in other contexts
sns.jointplot(x = "total_bill", y = "tip", data = tips);
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "reg");
```

```python
sns.pairplot(tips, x_vars = ["total_bill", "size"], y_vars = ["tip"],
             size = 5, aspect = .8, kind = "reg");
```

```python
sns.pairplot(tips, x_vars = ["total_bill", "size"], y_vars = ["tip"],
             hue = "smoker", size = 5, aspect = .8, kind = "reg");
```

### Section 3: Plotting with Categorical Data 

In above codes, we mainly concern the continuous variables.

Here, our aim is to explore some discrete variables.

#### Section 3.1: Categorical scatterplots

Here, one variable is discrete, one variable is continuous values.

```python
sns.stripplot(x = "day", y = "total_bill", data = tips);
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True);
```

```python
# avoiding overlapping points 
sns.swarmplot(x = "day", y = "total_bill", data = tips);
```

Here, add one discrete variable - sex.

There are two discrete variables and one continuous variable.

```python
sns.swarmplot(x = "day", y = "total_bill", hue = "sex", data = tips);
```
