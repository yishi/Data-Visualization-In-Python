### 6 Relationship between multiple variables  多个变量之间的关系

#### Face Plot 脸谱图

- 将多个变量一次性的展现在人脸的各个部位上，如耳朵、头发、眼睛、鼻子等。
- 较大的数值会以较多的头发或更大的眼睛来表示，较小的数值则会对面部特征进行收缩。
- 图中为2008赛季NBA前50位球员的统计数据。每一位球员都会用一个脸谱来表示。


```python
Image("img/face.png")
```

#### Rose Plot 玫瑰图（极坐标图）

- 图中为美国各州犯罪情况。
- 每个州用一朵玫瑰表示。
- 每一个花瓣代表一种犯罪类型（如抢劫、故意伤害、入室盗窃等）。
- 花瓣较长则代表该犯罪类型数量较高。


```python
Image("img/rose.png")
```

#### Heatmap 热力图

- 图中为2008赛季NBA球员的统计数据。
- 每一行代表一个球员的各项指标的高低，
- 每一列代表各个球员的某一项统计指标（如PTS：场均得分，FGP：投篮命中率），
- 横轴为一系列统计指标，纵轴为球员姓名，颜色的深浅为指标数值的大小。


```python
Image("img/heat.jpg")
```

#### Parallel Coordinate Plot 平行坐标图

- 方便描述群组或各变量之间的关系。
- 代表各变量的多条轴平行放置，每个对象都有一根线条，可在多个对象间寻找共同的变化趋势。
- 图中为美国国家教育统计中心的数据。
- 横轴为一系列统计指标（各州高中生在阅读、数学和写作各科目的SAT平均得分、参加考试人数的百分比、高中毕业生就业率和高中生辍学率），纵轴代表每个变量的取值范围。


```python
Image("img/parallel.png")
```
